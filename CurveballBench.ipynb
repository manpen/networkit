```python
import networkit as nk
import time
```

```python
G = nk.graphio.EdgeListReader(" ", 0, directed=True).read("/home/manpen/code/data/netflix/netflix10000.txt")
nk.overview(G)
```

```python
nLeft = 0
while G.degreeOut(nLeft) > 0:
    nLeft += 1
nRight = G.numberOfNodes() - nLeft    
nLeft, nRight
```

```python
alg = nk.randomization.Curveball(G)

trades = nk.randomization.CurveballUniformTradeGenerator(10 * nRight, (nLeft, nLeft+nRight))
start = time.time()
alg.run(trades)
G2 = alg.getGraph()
elapsed = time.time() - start

alg.elapsedMilliseconds(), elapsed
```
